```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Stars.csv')
df.head()
```

```python
df.info()
```

```python
class labelencoder:
  def __init__(self,columns):
    self.columns = columns

  def fit_transform(self,df):
    from sklearn.preprocessing import LabelEncoder
    for col in self.columns:
      le = LabelEncoder()
      df[col] = le.fit_transform(df[col])
    return df
```

```python
df = labelencoder(['Star category','Spectral Class','Star color']).fit_transform(df)
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns = ['Spectral Class'])
y = df['Spectral Class']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 2529)
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(x_train,y_train)
```

```python
pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test,pred)*100
score
```
